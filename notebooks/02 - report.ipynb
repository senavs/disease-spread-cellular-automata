```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../outputs/8f9498c8-7b1b-4473-afbf-5645aee49b08/reports/raw.csv', sep=";")
df.drop('id', axis=1, inplace=True)
df.head()
```

# Filtered DF

```python
counted_subject_state = df.iloc[:,1:].apply(pd.value_counts).fillna(0)
```

# Infection

```python
x = counted_subject_state.columns
y = counted_subject_state.loc['INFECTIOUS']

fig, ax = plt.subplots(1, 1, figsize=(16, 8))
plt.plot(x, y)
```

```python
x = counted_subject_state.columns
y = counted_subject_state.loc['DEAD']

fig, ax = plt.subplots(1, 1, figsize=(16, 8))
plt.plot(x, y)
```
